# Model1

```python
import pandas as pd
from pycaret.regression import *

model_1 = load_model('task1_inclued_cysc')#task1_inclued_cysc.pkl文件为Model1训练好的二进制文件

ID = ['ID','householdID']
bio = ['sysbp','diabp','pulse','wbc','mcv','plt','bun','glu','crea','cho','tg','hdl','ldl','crp','hbalc','ua','htc','hgb','cysc']
chronic = ['hibpe','diabe','cancre','lunge','hearte','stroke','psyche','arthre','dyslipe','livere','kidneye','digeste','asthmae','memrye']
limit = ['dressa','batha','eata','beda','toilta','urina','moneya','medsa','shopa','mealsa','housewka','joga','walk1kma','walk100a','chaira','climsa','stoopa','lifta','dimea','armsa']
cd = ['cesd10','shlta','slfmem','imrc','dlrc','ser7','orient','draw']
feature_1 = ID+bio+chronic+limit+cd #除了 ID = ['ID','householdID'], 其他的都是需要在页面上展示的变量

d3 = pd.read_stata('2015.dta')

d3 = d3[(d3.age >= 40) &(d3.age <= 85)]
d4 = d3[feature_1+['age']]
df_2015 = d4.dropna(subset=['age'])

pred = predict_model(model_1,data=df_2015)
    
pred = pred.reset_index(drop=True)
pred
```

pred中的Label列即为预测值，在网页展示时只需每次读取一条数据，把最后面的Label值传到网页前端即可。

# Model2

```python
import pandas as pd
model_2 = load_model('task2_include_cysc') #task2_inclued_cysc.pkl文件为Model1训练好的二进制文件

d3 = pd.read_stata('2011.dta')

d3 = d3[(d3.age >= 40) &(d3.age <= 85)]

feature_2 = ['ID','householdID','wbc','mcv','plt','bun','glu','crea','cho','tg','hdl','ldl','crp','hbalc','ua','htc','hgb','cysc']

d4 = d3[feature_2+['age']]
df_2011_numeric = d4.dropna(subset=['age'])

pred = predict_model(model_2,data=df_2011_numeric)

pred = pred.reset_index(drop=True)
pred
```

Model2的变量相对Model1更少，同样的， pred的最后一列Label代表预测值，网页每次只需读取一条输入数据，再将最后的Label值传到前端即可。
