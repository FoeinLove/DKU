```python
import pandas as pd
```

```python
d1 = pd.read_csv('dat_wl.csv',index_col=0)
```

```python
d1 = d1.drop_duplicates().dropna()
```

```python
from pycaret.regression import *
reg = setup(d1,target='number',ignore_features=['datetime','day','sunrise','sunset','size']
                                                ,normalize_method = 'minmax'
                                                ,normalize=True
                                                ,ignore_low_variance=True,remove_multicollinearity=True
                                                ,session_id=123)
```

```python
 best = compare_models()
```
